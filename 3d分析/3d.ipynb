```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C://Users/Jack/DATA/3dfix.csv')
```

```python
data2=data[['开奖日期','1','3','6']]
```

```python
plt.plot(data2['1'])
plt.show()
```

```python
b=data2['1']
```

```python
ts_log=np.log(b)
ts_diff=ts_log.diff(1)
ts_diff=ts_diff.dropna()
plt.plot(ts_diff)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(ts_diff)
plot_pacf(ts_diff)
```

```python
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(ts_log,order=(2,1,2))
result_ARIMA=model.fit(disp=-1)
plt.plot(ts_log_diff,color='#ffff00')
plt.plot(result_ARIMA.fittedvalues,color='#000ff')
```
